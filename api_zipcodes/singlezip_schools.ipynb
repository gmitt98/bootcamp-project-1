```python
# Dependencies
import requests
import json
from pprint import pprint
import pandas as pd
import hvplot.pandas

# Import the API key
from config import geoapify_key
```

```python
zipcode = "34429" #insert zipcode of choice

params = {
    "apiKey": geoapify_key,
    "country":"United States",
    "postcode": zipcode
}
# Build URL using the geocode endpoint
base_url = f"https://api.geoapify.com/v1/geocode/search?text={zipcode}&lang=en&type=postcode&filter=countrycode:us"
```

```python
#run request
response = requests.get(base_url, params=params).json()
```

```python
 # Extract lat/lon
lat = response["features"][0]["properties"]["lat"]
lon = response["features"][0]["properties"]["lon"]
place_id = response["features"][0]["properties"]["place_id"]
```

```python
# Set paramaters for type of place
categories = "education.school"

# Set parameters for type of search
bias = f"proximity:{lon},{lat}"
filters = f"place:{place_id}"
```

```python
#parameters dictionary
params = {
    "categories": categories,
    "bias": bias,
    "filters":filters,
    "apiKey": geoapify_key    
}
```

```python
# Set base URL
base_url = f"https://api.geoapify.com/v2/places?"
```

```python
response = requests.get(base_url, params=params).json()
```

```python
# Extract the features list from the dictionary
features_list = response['features']

# Extract the relevant properties from each feature
rows = []
for feature in features_list:
    properties = feature['properties']
    rows.append(properties)

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(rows)
df
```

```python
df = df.dropna(subset=['name', 'postcode'])
df = df.drop(columns=['country','country_code','state','street','state_code','formatted','address_line1',
                     'address_line2','categories','details','datasource','place_id','housenumber','district','suburb',
                      'quarter','old_name','city','hamlet'],errors='ignore')
df = df.rename(columns={'postcode': 'zipcode'})
df
```

```python
area_df = df
zip_df = area_df.loc[area_df['zipcode'] == zipcode]
```

```python
#read in total market data to get HPIs and compare zip code schools data
from pathlib import Path
total_market_csv = Path('../data/cleaned data/total_market_data_merged.csv')
HPI_school_market = pd.read_csv(total_market_csv)
HPI_school_market .rename(columns = {'zip_code':'zipcode'}, inplace = True)
HPI_school_market['zipcode'] = HPI_school_market ['zipcode'].apply(str)
zip_hpi_df = pd.merge(area_df, HPI_school_market, on='zipcode', how="left")
zip_hpi_df.drop(columns=['RECESSION_FLAG', 'avg_rate_for_year', 'Median_hh_income', 'Year_Avg_Unempl', 'state', 'bucketed_year', 'State_Num', 
                  'Percent Change in Resident Population'], axis=1, inplace=True)
zip_hpi_df.head()
```

# Visualizations of zip code data

```python
#zip code 97214 plotting schools and HPI values for Portland, OR area data 
zip_hpi_2022 = zip_hpi_df[zip_hpi_df['year']==2022]
zip_hpi_2022.hvplot.points('lon', 'lat', geo=True, frame_width = 800, frame_height = 600, hover_cols=['name', 'zipcode', 'HPI'],
                                 tiles='OSM', color='blue', title = "Schools and HPI trends in Multnomah County")
```

```python
#average HPI in Oregon zip code 97214:
print(f"The average HPI in this area of Oregon (near zip 97214) is {zip_hpi_2022['HPI'].mean()}")
print(f"There were {zip_hpi_2022['name'].count()} data points for this area.")
```
